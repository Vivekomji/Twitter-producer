```python
result = dict()
for i in data_lst:
#     print(i['Donor'], i['TOTAL ACT-A funding'])
    r1 = i['Donor']
    r2 = i['TOTAL ACT-A funding']
    r2 = r2[1:]
    r2 = r2[:-8]
    r2 = float(r2)
#     print(type(r2))
    result[r1] = r2
    
```

```python
def make_barplot_using_seaborn(response_data):
    plt.figure(figsize=(18,18))
    sns.barplot(list(response_data.keys()), list(response_data.values()))
    plt.show()
```

```python
def make_wordcloud(response_data):
    wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(response_data)
    plt.figure(figsize=(15,8))
    plt.imshow(wordcloud)
```

```python
def make_barplot_using_matplotlib(response_data):
    plt.figure(figsize=(18,18))
    plt.bar(*zip(*response_data.items()))
    plt.show()
```

```python
make_barplot_using_seaborn(result)
```

```python
make_wordcloud(result)
```

```python
make_barplot_using_matplotlib(result)
```

```python
def make_graph(my_dict):
    plt.figure(figsize=(18,18))
    myList = my_dict.items()
    myList = sorted(myList) 
    x, y = zip(*myList) 

    plt.plot(x, y)
    plt.show()
```

```python
make_graph(result)
```

```python
def plot_pie_chart(your_data):
    plt.figure(figsize=(18,18))
    plt.pie([float(your_data[v]) for v in your_data], labels=[str(k) for k in your_data], autopct='%1.1f%%')
    plt.show()
```

```python
plot_pie_chart(result)
```
