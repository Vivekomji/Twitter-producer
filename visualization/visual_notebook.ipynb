```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

```python
print(list_word_count)
```

```python
def make_dictionary(li):
    dic=dict()
    for k,v in li:
        dic[k]=int(v)
    return dic
```

```python
word_count = make_dictionary(list_word_count)
```

```python
print(word_count)
```

```python
def make_barplot_using_seaborn(word_count):
    plt.figure(figsize=(18,18))
    sns.barplot(list(word_count.keys()), list(word_count.values()))
    plt.show()
```

```python
def make_wordcloud(word_count):
    wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(word_count)
    plt.figure(figsize=(15,8))
    plt.imshow(wordcloud)
```

```python
def make_barplot_using_matplotlib(word_count):
    plt.figure(figsize=(18,18))
    plt.bar(*zip(*word_count.items()))
    plt.show()
```

```python
make_barplot_using_seaborn(word_count)
```

```python
make_wordcloud(word_count)
```

```python
make_barplot_using_matplotlib(word_count)
```

```python
def make_graph(my_dict):
    plt.figure(figsize=(18,18))
    myList = my_dict.items()
    myList = sorted(myList) 
    x, y = zip(*myList) 

    plt.plot(x, y)
    plt.show()
```

```python
make_graph(word_count)
```

```python
def plot_pie_chart(your_data):
    plt.figure(figsize=(18,18))
    plt.pie([float(your_data[v]) for v in your_data], labels=[str(k) for k in your_data], autopct='%1.1f%%')
    plt.show()
```

```python
plot_pie_chart(word_count)
```
